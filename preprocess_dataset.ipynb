## ISIC2018

```python
#create dataset dirctory
import os

class_list = ['AKIEC', 'BCC', 'BKL', 'DF', 'MEL', 'NV', 'VASC']
set_list = ['./training_set', './test_set', './validation_set']
for i in set_list:
    if not os.path.isdir(i):
        os.mkdir(i)
        for j in class_list:
            os.mkdir(i + '/' + j)
```

```python
# split training set, validation set, test set
import os
import pandas as pd
import math
import random

# 6:1:3
train_rate, test_rate = 0.6, 0.3

space_path = os.getcwd()
source_path_prefix = space_path + '/ISIC2018/'
train_path_prefix = space_path + '/training_set/'
val_path_prefix = space_path + '/validation_set/'
test_path_prefix = space_path + '/test_set/'

df = pd.read_csv(source_path_prefix + 'ISIC2018_Task3_Training_GroundTruth.csv')

clas = {}
MEL = df[df['MEL'] == 1]['image'].tolist()
NV = df[df['NV'] == 1]['image'].tolist()
BCC = df[df['BCC'] == 1]['image'].tolist()
AKIEC = df[df['AKIEC'] == 1]['image'].tolist()
BKL = df[df['BKL'] == 1]['image'].tolist()
DF = df[df['DF'] == 1]['image'].tolist()
VASC = df[df['VASC'] == 1]['image'].tolist()

print(len(MEL),len(NV),len(BCC),len(AKIEC),len(BKL),len(DF),len(VASC))

clas["MEL"] = MEL
clas["NV"] = NV
clas["BCC"] = BCC
clas["AKIEC"] = AKIEC
clas["BKL"] = BKL
clas["DF"] = DF
clas["VASC"] = VASC

random.seed(123)
for key, value in clas.items():
    random.shuffle(value)
    train_list = value[:math.ceil(len(value)*train_rate)]
    test_list = value[-math.floor(len(value)*test_rate):]
    val_list = [i for i in value if ((i not in train_list) and (i not in test_list))]
    print("{}: {} -> {} train, {} test, {} val".format(key, len(value), len(train_list), len(test_list), len(val_list)))
    for v in train_list:
        source_path = source_path_prefix + str(v) + '.jpg'
        if os.path.exists(source_path):
            train_path = train_path_prefix + str(key) + '/' + str(v) + '.jpg'
            os.rename(source_path, train_path)
    for v in test_list:
        source_path = source_path_prefix + str(v) + '.jpg'
        if os.path.exists(source_path):
            test_path = test_path_prefix + str(key) + '/' + str(v) + '.jpg'
            os.rename(source_path, test_path)
    for v in val_list:
        source_path = source_path_prefix + str(v) + '.jpg'
        if os.path.exists(source_path):
            val_path = val_path_prefix + str(key) + '/' + str(v) + '.jpg'
            os.rename(source_path, val_path)
```

## Kvasir

```python
#create dataset dirctory
import os

class_list = ['dyed-lifted-polyps', 'dyed-resection-margins', 'esophagitis', 'normal-cecum', 'normal-pylorus', 'normal-z-line', 'polyps', 'ulcerative-colitis']
set_list = ['./training_set_2', './test_set_2', './validation_set_2']
for i in set_list:
    if not os.path.isdir(i):
        os.mkdir(i)
        for j in class_list:
            os.mkdir(i + '/' + j)
```

```python
# split training set, validation set, test set
import os
import pandas as pd
import math
import random

# 1:1, 2fold
train_rate, test_rate = 0.5, 0.5

space_path = os.getcwd()
source_path_prefix = space_path + '/Kvasir/'
train_path_prefix = space_path + '/training_set_2/'
val_path_prefix = space_path + '/validation_set_2/'
test_path_prefix = space_path + '/test_set_2/'


clasnames = []
for _, dirs, _ in os.walk(source_path_prefix):
    clasnames = dirs
    break

clas={}
for c in clasnames:
    classpath = source_path_prefix + c
    # print(classpath)
    for _, _, files in os.walk(classpath):
        # print([classpath + i for i in files])
        clas[c] = [classpath + '/' + i for i in files]


random.seed(123)
for key, value in clas.items():
    random.shuffle(value)
    train_list = value[:math.ceil(len(value)*train_rate)]
    test_list = value[-math.floor(len(value)*test_rate):]
    val_list = [i for i in value if ((i not in train_list) and (i not in test_list))]
    print("{}: {} -> {} train, {} test, {} val".format(key, len(value), len(train_list), len(test_list), len(val_list)))
    for v in train_list:
        if os.path.exists(v):
            train_path = train_path_prefix + str(key) + '/' + str(v.split('/')[-1])
            os.rename(v, train_path)
    for v in test_list:
        if os.path.exists(v):
            test_path = test_path_prefix + str(key) + '/' + str(v.split('/')[-1])
            os.rename(v, test_path)
    for v in val_list:
        if os.path.exists(v):
            val_path = val_path_prefix + str(key) + '/' + str(v.split('/')[-1])
            os.rename(v, val_path)
```
